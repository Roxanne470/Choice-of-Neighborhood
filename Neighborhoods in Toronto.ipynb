```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
```

```python
url = "https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M"
url
```

```python
html = urlopen(url)
```

```python
soup = BeautifulSoup(html, 'lxml')
type(soup)
```

```python
rows = soup.find_all('tr')
```

```python
import re
list_rows = []
for row in rows:
    cells = row.find_all('td')
    str_cells = str(cells)
    clean = re.compile('<.*?>')
    clean2 = (re.sub(clean, '', str_cells))
    list_rows.append(clean2)
```

```python
df = pd.DataFrame(list_rows)
```

```python
df.head()
```

```python
df1 = df[0].str.split(',', expand=True)
df1.head(5)
```

```python
df1[0] = df1[0].str.strip('[')
df1.head()
```

```python
range(df1.shape[1])
```

```python
for i in range(df1.shape[1]):
    df1[i] = df1[i].str.strip(']')
    df1[i] = df1[i].str.strip('\n')
```

```python
df1.rename(columns = {0: 'PostalCode', 1: 'Borough'}, inplace = True)
df1.head()
```

```python
df2 = df1[df1['Borough'] != ' Not assigned']
```

```python
df2 = df2[1:]
```

```python
df2.reset_index(drop = True, inplace = True)
```

```python
df2['Neighborhood'] = df2[df2.columns[2:]].apply(lambda x: ','.join(x.dropna().astype(str)), axis = 1)
df2.head()
```

```python
data = df2[['PostalCode','Borough','Neighborhood']]
data.head()
```
